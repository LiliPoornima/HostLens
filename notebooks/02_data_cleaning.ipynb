```python
import pandas as pd

listings = pd.read_csv("../data/raw/listings.csv")
calendar = pd.read_csv("../data/raw/calendar.csv")
reviews = pd.read_csv("../data/raw/reviews.csv")
neighbourhoods = pd.read_csv("../data/raw/neighbourhoods.csv")
```

```python
# Section 3.2: Data Cleaning & Standardization
# Task: Clean the price column

listings["price"] = (
    listings["price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
# Section 3.2: Data Cleaning & Standardization
# Task: Standardize date fields

listings["last_scraped"] = pd.to_datetime(listings["last_scraped"])
calendar["date"] = pd.to_datetime(calendar["date"])
reviews["date"] = pd.to_datetime(reviews["date"])
```

```python
# Section 3.2: Data Cleaning & Standardization
# Task: Save cleaned datasets

listings.to_csv(
    "../data/processed/listings_cleaned.csv",
    index=False
)

#calendar.to_csv(
#    "../data/processed/calendar_cleaned.csv",
#    index=False
#)

reviews.to_csv(
    "../data/processed/reviews_cleaned.csv",
    index=False
)
```

```python
# Section 3.2: Data Cleaning & Standardization
# Task: Normalize categorical fields

listings["room_type"] = (
    listings["room_type"]
    .str.strip()
    .str.title()
)

listings["property_type"] = (
    listings["property_type"]
    .str.strip()
    .str.title()
)
```

```python
# Section 3.2: Data Cleaning & Standardization
# Task: Standardize date fields

listings["last_scraped"] = pd.to_datetime(
    listings["last_scraped"]
)

reviews["date"] = pd.to_datetime(
    reviews["date"]
)
```

```python
# Section 3.2: Data Cleaning & Standardization
# Task: Standardize geographic fields

listings["neighbourhood_cleansed"] = (
    listings["neighbourhood_cleansed"]
    .str.strip()
    .str.title()
)

listings["latitude"] = (
    listings["latitude"]
    .round(6)
)

listings["longitude"] = (
    listings["longitude"]
    .round(6)
)
```

# Data Cleaning and Standardization

This notebook performs data cleaning and standardization on the Inside Airbnb dataset for New York City.

## Objectives

- Standardize price columns by removing currency symbols, handling separators, and converting values to numeric types.
- Parse and standardize date fields across all datasets.
- Normalize free-text fields such as room types and property types.
- Assess and handle missing values using appropriate strategies.
- Validate records against domain rules and document data quality issues.
- Standardize geographic fields, including neighbourhood names and coordinate precision.
- Generate clean, analytics-ready datasets for downstream analysis and modeling.

## Datasets Used

- `listings.csv`
- `calendar.csv`
- `reviews.csv`
- `neighbourhoods.csv`
