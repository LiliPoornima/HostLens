# Section Summary: Data Modeling

## Goal

The goal of this notebook is to design and implement a dimensional model for analytical queries on Airbnb data.

## Objectives

- Design a star schema.
- Define fact and dimension tables.
- Create relationships between entities.
- Support analytical and business intelligence queries.

## Star Schema Design

### Fact Table

#### fact_listings

Stores measures and metrics related to Airbnb listings:

- listing_id
- host_id
- price
- number_of_reviews
- review_scores_rating
- reviews_per_month
- total_reviews

### Dimension Tables

#### dim_hosts

- host_id
- host_name
- host_is_superhost
- host_location

#### dim_neighbourhoods

- neighbourhood_cleansed
- neighbourhood_group_cleansed

#### dim_property

- property_type
- room_type
- bedrooms
- beds

#### dim_reviews

- listing_id
- first_review_y
- last_review_y
