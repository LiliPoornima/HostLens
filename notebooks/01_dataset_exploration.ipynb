```python
# Section 3.1: Data Ingestion & Profiling
# Task: Validate business rules

negative_prices = (listings["price"] < 0).sum()

print(f"Listings with negative prices: {negative_prices}")
```

```python
# Section 3.1: Data Ingestion & Profiling
# Task: Inspect the price column

print(listings["price"].head())
print(listings["price"].dtype)
```

```python
# Section 3.2: Data Cleaning & Standardization
# Task: Standardize the price column

listings["price"] = (
    listings["price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

print(listings["price"].head())
print(listings["price"].dtype)
```

```python
# Section 3.1: Data Ingestion & Profiling
# Task: Validate business rules (rerun after cleaning)

negative_prices = (listings["price"] < 0).sum()

print(f"Listings with negative prices: {negative_prices}")
```

```python
# Section 3.2: Data Cleaning & Standardization
# Task: Standardize date fields

listings["last_scraped"] = pd.to_datetime(listings["last_scraped"])
calendar["date"] = pd.to_datetime(calendar["date"])
reviews["date"] = pd.to_datetime(reviews["date"])
```

```python
# Section 3.2: Data Cleaning & Standardization
# Task: Verify date conversions

print(listings["last_scraped"].dtype)
print(calendar["date"].dtype)
print(reviews["date"].dtype)
```
